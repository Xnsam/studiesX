Two nums

```python
def two_sums(nums: str, target: int) -> list:
    seen = {}
    for idx, v in enumerate(nums):
        diff = target - v
        if diff in seen:
            return [seen[diff], idx]
        
        seen[v] = idx
    
    return [0, 0]

nums = [2, 7, 11, 15]
target = 9

two_sums(nums, target)
```

Buy and sell

```python
def buy_sell(prices: list) -> int:
    l, r = 0, 1
    max_p = 0

    while r < len(prices):
        if prices[r] > prices[l]:
            profit = prices[r] - prices[l]
            max_p = max(max_p, profit)
        else:
            l = r
            
        r += 1

    return max_p

arr = [7,1,5,3,6,4]

buy_sell(arr)
```

TreeConstructor

```python
def tree_constructor(str_arr: list) -> bool:
    parent, child = {}, {}

    for pair in str_arr:
        c, p = map(int, pair.strip("()").split(","))
        parent[p] = parent.get(p, 0) + 1
        child[c] = child.get(c, 0) + 1
        print(parent, child)
        if parent[p] > 2 or child[c] > 1:
            return False
    return True


str_arr = ["(1, 2)", "(2, 4)", "(5, 7)"]
str_arr = ["(1,2)", "(3,2)", "(5,2)"] 

tree_constructor(str_arr)
```

```python
import numpy as np

# A precise value (like a neural network weight)
original_value = 0.1234567890123456789

# 1. High Precision (64-bit / 8 bytes)
# Can store up to ~15-17 decimal digits
fp64 = np.float64(original_value)

# 2. Standard Precision (32-bit / 4 bytes)
# Can store up to ~7 decimal digits
fp32 = np.float32(original_value)

# 3. Half Precision (16-bit / 2 bytes)
# Only stores ~3-4 decimal digits accurately
fp16 = np.float16(original_value)

print(f"Original: {original_value}")
print(f"64-bit:   {fp64}")
print(f"32-bit:   {fp32}")
print(f"16-bit:   {fp16}")

# Show the error (The "Cost" of quantization)
print(f"\nError in 16-bit: {abs(fp64 - fp16):.10f}")
```

```python
s = "aaabbbcca"
output = "a3b3c2a1"

def convert_char(s : str) -> dict:
    final_str = ""
    stack = []
    for c in s:
        if not stack:
            stack.append(c)
        elif c in stack:
            stack.append(c)
        else:
            final_str += f"{c}{len(stack)}"
            stack = []

    return final_str

convert_char(s)
```
